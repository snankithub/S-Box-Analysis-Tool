```python
import numpy as np
from tabulate import tabulate

# Function to compute the Difference Distribution Table (DDT)
def compute_ddt(sbox):
    n = 8  # For an 8-bit S-box
    m = 1 << n  # m = 2^n = 256
    ddt = np.zeros((m, m), dtype=int)
    for x in range(m):
        for dx in range(m):
            dy = sbox[x] ^ sbox[x ^ dx]
            ddt[dx][dy] += 1
    return ddt

# Function to compute the Linear Approximation Table (LAT)
def compute_lat(sbox):
    n = 8  # For an 8-bit S-box
    m = 1 << n  # m = 2^n = 256
    lat = np.zeros((m, m), dtype=int)
    for a in range(m):
        for b in range(m):
            bias = 0
            for x in range(m):
                if bin(a & x).count('1') % 2 == bin(b & sbox[x]).count('1') % 2:
                    bias += 1
            lat[a][b] = bias - (m // 2)
    return lat

# Function to analyze the S-box
def sbox_analysis(sbox):
    ddt = compute_ddt(sbox)
    lat = compute_lat(sbox)
    return {
        "DDT": ddt,
        "LAT": lat,
    }

# Function to pretty-print tables
def print_table(table, title):
    print(f"\n{title}\n")
    print(tabulate(table, headers="keys", tablefmt="grid"))

# Example 8-bit S-box (256 entries)
sbox = [
    0x08, 0xDB, 0xA6, 0xA3, 0x6A, 0x3B, 0x07, 0xD9, 0x57, 0xED, 0xD4, 0xAC, 0x95, 0xC5, 0x78, 0xE8,
    0x51, 0x4E, 0x2C, 0x75, 0x28, 0x1C, 0x8D, 0x44, 0xC2, 0x4F, 0x59, 0xBC, 0x6F, 0x79, 0xBD, 0x3D,
    0x89, 0xD0, 0x34, 0x0C, 0xC7, 0x18, 0x30, 0x32, 0xCE, 0x66, 0x4A, 0x7C, 0xC0, 0xA4, 0x8C, 0x2D,
    0x01, 0x45, 0x86, 0x98, 0x10, 0x04, 0xEE, 0x1A, 0xF6, 0xD1, 0x20, 0x8B, 0x02, 0x15, 0xFD, 0xB2,
    0x03, 0x39, 0x2E, 0xAE, 0x47, 0x81, 0x53, 0xE6, 0xD8, 0x17, 0x0F, 0x70, 0xA1, 0x7F, 0xA8, 0x58,
    0x36, 0xB4, 0x13, 0x16, 0x7E, 0x0D, 0xC3, 0x23, 0x80, 0x8E, 0xF9, 0x27, 0xE7, 0x5B, 0x5F, 0x22,
    0xF7, 0x99, 0x82, 0xB3, 0x87, 0xE5, 0x7D, 0xC6, 0x0A, 0xE2, 0xEF, 0xC1, 0xD6, 0x40, 0xF4, 0xB8,
    0xC4, 0xFB, 0x49, 0xEA, 0x62, 0x1D, 0x2B, 0x50, 0x84, 0xF5, 0x55, 0x64, 0xAB, 0x69, 0xF0, 0x4C,
    0x26, 0xCB, 0x41, 0x5E, 0xE1, 0x2F, 0x8F, 0xBF, 0xFF, 0xAD, 0x83, 0x3E, 0x9B, 0xB0, 0x8A, 0x56,
    0x37, 0x4D, 0x24, 0x6B, 0x9D, 0x9A, 0x48, 0xBB, 0x7B, 0x1B, 0x68, 0x74, 0x21, 0x12, 0xD7, 0x46,
    0x6C, 0x76, 0x4B, 0x09, 0x29, 0xCF, 0xBA, 0xDD, 0xD2, 0x3A, 0xAA, 0x67, 0xC8, 0x05, 0x52, 0x91,
    0xFA, 0x88, 0x9F, 0x1E, 0xAF, 0xF2, 0x77, 0xB1, 0x9E, 0xF8, 0x6D, 0xA9, 0x5D, 0x71, 0xA0, 0x2A,
    0x3C, 0xB9, 0x19, 0xC9, 0x5A, 0xEB, 0x63, 0xDF, 0xD5, 0xA2, 0x11, 0xFE, 0x54, 0x85, 0xF3, 0xCA,
    0x93, 0xF1, 0xEC, 0x7A, 0xE0, 0xB5, 0x1F, 0x72, 0x61, 0x6E, 0x73, 0xB7, 0xA7, 0xFC, 0x9C, 0x90,
    0x92, 0x43, 0xE9, 0x96, 0x14, 0x42, 0x65, 0xD3, 0x38, 0xCC, 0xDC, 0x0E, 0x5C, 0x31, 0x35, 0xBE,
    0xDA, 0xCD, 0xDE, 0x97, 0x3F, 0x94, 0x00, 0x33, 0x25, 0xA5, 0xB6, 0xE3, 0x0B, 0xE4, 0x06, 0x60
]

# Run the analysis
results = sbox_analysis(sbox)

# Print the results
print_table(results["DDT"], "Difference Distribution Table (DDT)")
print_table(results["LAT"], "Linear Approximation Table (LAT)")
```
